1. What is the difference between a function and a method in Python?
 - A function is an independent block of reusable code, while a method is a function associated with an object and operates on its data.
 - Function Example:
    def add(a, b): return a + b  # Independent function
    print(add(2, 3))  # Output: 5
  - Method Example:
  class Math:  
  def add(self, a, b): return a + b  # Method inside a class
  m = Math()
  print(m.add(2, 3))  # Output: 5


2. Explain the concept of function arguments and parameters in Python?
 - Parameters are variables in a function definition, while arguments are actual values passed to the function when calling it.
 - Example: def greet(name):  # 'name' is a parameter
    return f"Hello, {name}!"
    print(greet("Ryan"))  # "Ryan" is an argument, Output: Hello, Ryan!

4. What is the purpose of the `return` statement in a Python function?
 - The return statement in Python functions is used to send a result back to the caller and terminate the function execution.
 - Example:
def square(num):  
    return num * num  
print(square(4))  # Output: 16
If return is omitted, the function returns None by default.


5. What are iterators in Python and how do they differ from iterables?
  - An iterator is an object that represents a stream of data and can be iterated using next(), whereas an iterable is an object that can return an iterator using iter().
  - Example:
  nums = [1, 2, 3]  # Iterable (list)
  it = iter(nums)  # Iterator
  print(next(it))  # Output: 1
  print(next(it))  # Output: 2
  - Iterables include lists, tuples, and dictionaries, while iterators fetch values one at a time using next().


6.  Explain the concept of generators in Python and how they are defined?
  - A generator in Python is a special type of iterator that yields values one at a time using yield, saving memory compared to regular functions.
  - Example:
  def count_up(n):  
    for i in range(n):  
        yield i  
    gen = count_up(3)  
    print(next(gen))  # Output: 0  
    print(next(gen))  # Output: 1  
   - Generators are defined using functions with yield, and they pause execution between calls.


7.  What are the advantages of using generators over regular functions?
  - Generators are memory-efficient, as they yield values one at a time instead of storing them all in memory. They allow lazy evaluation, making them useful for large datasets and infinite sequences.
  - Example:
  def count_up(n):  
    for i in range(n):  
        yield i  
    gen = count_up(3)  
    print(next(gen))  # Output: 0  
   - Unlike regular functions, generators do not store the entire output, reducing memory usage.


8. What is a lambda function in Python and when is it typically used?
  - A lambda function is an anonymous, one-liner function defined using the lambda keyword, often used for short, simple operations.
  - Example:
  square = lambda x: x * x  
  print(square(4))  # Output: 16  
  - It is typically used for quick, throwaway functions in map(), filter(), and sorting operations.


9.  Explain the purpose and usage of the `map()` function in Python?
  - The map() function applies a given function to all items in an iterable and returns a map object (iterator).
  - Example:
  nums = [1, 2, 3]  
  squares = list(map(lambda x: x * x, nums))  
  print(squares)  # Output: [1, 4, 9]  
  - It is useful for transforming lists without using explicit loops.


10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  - 1. map() = Applies a function to each item in an iterable and returns a new iterable.
  Example:
  nums = [1, 2, 3]
  squares = list(map(lambda x: x * x, nums))
  print(squares)  # Output: [1, 4, 9]
  - 2. filter() = Filters elements based on a condition and returns only matching items.
  Example:
  evens = list(filter(lambda x: x % 2 == 0, nums))
   print(evens)  # Output: [2]
   - 3. reduce() =  Applies a function cumulatively to elements, reducing them to a single value.
   Example:
   from functools import reduce
   total = reduce(lambda x, y: x + y, nums)
  print(total)  # Output: 6
   - map() transforms, filter() selects, and reduce() aggregates values.

```python
1. #Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list?
def sum_of_evens(nums):  
    return sum(filter(lambda x: x % 2 == 0, nums))  

print(sum_of_evens([1, 2, 3, 4, 5, 6])) 
```

```python
3. # Implement a Python function that takes a list of integers and returns a new list containing the squares of each number?
def square_numbers(nums):  
    return list(map(lambda x: x * x, nums))  

print(square_numbers([1, 2, 3, 4]))
```

```python
4.  #Write a Python function that checks if a given number is prime or not from 1 to 200?
def is_prime(n):  
    if n < 2: return False  
    return all(n % i != 0 for i in range(2, int(n**0.5) + 1))  

print([n for n in range(1, 201) if is_prime(n)])  
```

```python
5. #Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms?
class FibonacciIterator:  
    def __init__(self, n):  
        self.n, self.a, self.b, self.count = n, 0, 1, 0  

    def __iter__(self):  
        return self  

    def __next__(self):  
        if self.count >= self.n:  
            raise StopIteration  
        self.a, self.b = self.b, self.a + self.b  
        self.count += 1  
        return self.a  

fib = FibonacciIterator(10)  
print(list(fib))  
```

```python
6. #Write a generator function in Python that yields the powers of 2 up to a given exponent?
def power_of_two(exponent):  
    for i in range(exponent + 1):  
        yield 2 ** i  

gen = power_of_two(5)  
print(list(gen))
```

```python
7. #Implement a generator function that reads a file line by line and yields each line as a string?
def read_file_line_by_line(filename):  
    with open(filename, 'r') as file:
        for line in file:  
            yield line.strip()  
            for line in read_file_line_by_line("example.txt"):
                print(line)  
```

```python
8. #Use a lambda function in Python to sort a list of tuples based on the second element of each tuple?
tuples = [(1, 3), (4, 1), (2, 2)]  
sorted_tuples = sorted(tuples, key=lambda x: x[1])  
print(sorted_tuples)
```

```python
9. #Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit?
celsius = [0, 20, 30, 40]  
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))  
print(fahrenheit)
```

```python
10. #Create a Python program that uses `filter()` to remove all the vowels from a given string?
def remove_vowels(s):  
    return "".join(filter(lambda x: x.lower() not in "aeiou", s))  

print(remove_vowels("Hello World"))
```

```python
11. #Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €. Write a Python program using lambda and map?

orders = [
    (34587, "Learning Python, Mark Lutz", 4, 40.95),
    (98762, "Programming Python, Mark Lutz", 5, 56.80),
    (77226, "Head First Python, Paul Barry", 3, 32.95),
    (88112, "Einführung in Python3, Bernd Klein", 3, 24.99)
]
result = list(map(lambda x: (x[0], x[2] * x[3] + (10 if x[2] * x[3] < 100 else 0)), orders))
print(result)
```
